## 1. Cesta básica

```python
#teste - cesta básica

cesta_basica_sp = quandl.get("BCB/7493", 
                             start_date = "2015-01-01")
```

```python
cesta_basica_sp.head()
```

```python
fig = px.line(x = cesta_basica_sp.index,
              y= cesta_basica_sp["Value"],
              title = "Valor da cesta básica em SP",
              labels = {"x": "Data", "y": "Valor em R$"})

fig.show()
```
